```python
from BattlePage import BattlePage
from TrainSetupPage import TrainSetupPage

from selenium import webdriver

import time
```

```python
player_team = ("gyarados,DRAGON_BREATH,AQUA_TAIL,CRUNCH\n"
                "trevenant,SUCKER_PUNCH,SHADOW_BALL,FOUL_PLAY\n"
                "jellicent,HEX,BUBBLE_BEAM,SHADOW_BALL")
opp_team = ("slaking,YAWN,BODY_SLAM\n"
            "porygon-z,LOCK_ON,FRUSTRATION\n"
            "audino,POUND,HYPER_BEAM")
league = '1500 gobattleleague'
difficulty = '0' # 0 is novice 3 is champion
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(0.3)
```

```python
train_setup_page = TrainSetupPage(driver)
```

```python
train_setup_page.set_team(player_team)
train_setup_page.set_league(league)
train_setup_page.set_difficulty(difficulty)
train_setup_page.set_opp_team(opp_team)
```

```python
battle_page = train_setup_page.battle()
```

Battle!

```python
# sync up turns as closely as possible
# update loop will start right as countdown ends
while(battle_page.get_battle_phase() == 'countdown'):
    pass
    
while(True):
    phase = battle_page.get_battle_phase()
    
    # check if game has ended
    if phase == 'game_over_screen':
        break
        
    elif phase == 'suspend_charged_shield':
        battle_page.use_shield()
        #battle_page.close_shield()
        
    elif phase == 'suspend_charged_attack':
        battle_page.charge_move(100)

    elif phase == 'suspend_switch_self':
        party = battle_page.get_switch_window_party()
        
        # simple alg to choose next pokemon:
        # pick pokemon on the left
        battle_page.switch_fainted_poke(party[0][1])

    elif phase == 'neutral':
        
        if previous_phase != 'neutral':
            print(battle_page.get_time())
            print()
            print(battle_page.get_name())
            print(battle_page.get_cp())
            print(battle_page.get_poke_hp())
            print()
            print(battle_page.get_opp_name())
            print(battle_page.get_opp_cp())
            print(battle_page.get_opp_hp())
            print()
            print(battle_page.get_remaining_poke())
            print(battle_page.get_shields_remaining())
            print()
            print(battle_page.get_opp_remaining_poke())
            print(battle_page.get_opp_shields_remaining())
            print()
            print()
            
            
            
        # simple alg to choose a charged move
        # if one of them is ready, use it
        charged_moves = battle_page.get_charged_moves()
        if charged_moves[1][1]:
            battle_page.use_charged_move(1)
        elif charged_moves[0][1]:
            battle_page.use_charged_move(0)
            
        else:
            battle_page.use_fast_move()
        
    previous_phase = phase
    # one input per turn, doesn't allow to queued actions
    time.sleep(0.5)
```
